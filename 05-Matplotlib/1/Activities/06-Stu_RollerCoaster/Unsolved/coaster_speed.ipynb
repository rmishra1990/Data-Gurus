```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_axis = np.arange(0,121,10)
x_axis
```

```python
danger_drop = [9, 8, 90, 85, 80, 70, 70, 65, 55, 60, 70, 65, 50]

railGun = [75, 70, 60, 65, 60, 45, 55, 50, 40, 40, 35, 35, 30]
```

```python
# Use dots or other markers for your plots, and change their colors
plt.plot(x_axis, danger_drop, marker="o", color="blue",label="Danger Drop")
plt.plot(x_axis, railGun, marker="^", color="red",label="RailGun")
```

```python
plt.legend(handles =[danger_drop,railGun],loc ="best")
```
