```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_axis = np.arange(1,13,1)
x_axis
```

```python
#Avg weather temp
fahrenheit = [39, 42, 51, 62, 72, 82, 86, 84, 77, 65, 55, 44]
```

```python
#Conversion to celsius
Celsius =[(x-32)*.56 for x in fahrenheit]
Celsius
```

```python
# Each point on the sine chart is marked by a blue circle
fahrenheit, = plt.plot(x_axis, fahrenheit, marker ='o', color='blue', label="Sine")
# Each point on the cosine chart is marked by a red triangle
Celsius, = plt.plot(x_axis, Celsius, marker='^', color='red', label="Cosine")
```

```python
plt.legend(handles =[fahrenheit,Celsius],loc ="best")
```
